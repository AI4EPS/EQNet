```python
import pandas as pd
from tqdm import tqdm
import json
from datetime import datetime
```

```python
with open("ncedc_catalogs.txt") as fp:
    catalogs = fp.read().splitlines()
```

```python
# catalogs
```

```python
event_dict = {}
for i, line in enumerate(tqdm(catalogs)):
    if len(line) > 150:
        event_time = line.split()[0][:-2]
    elif len(line) < 100:
        event_id = line.split()[-1]
        # print(i, event_time, event_id)
        event_dict[event_time] = event_id
    else:
        continue
    # print(i, event_time, event_id)
    # print(len(line))
    # if i > 20:
    #     raise
    if event_id == "8370":
        print(i, line)
```

```python
event_dict
```

```python
with open("event_id.json", "w") as fp:
    json.dump(event_dict, fp, indent=2)
```
